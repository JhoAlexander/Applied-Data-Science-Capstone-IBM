Above coordinates are just plain numbers that can not give you any intuitive insights about where are those launch sites. If you are very good at geography, you can interpret those numbers directly in your mind. If not, that's fine too. Let's visualize those locations by pinning them on a map.

```python
# Agrupar para obtener una fila única por sitio de lanzamiento
unique_sites = launch_sites_df.groupby('Launch Site').first().reset_index()

# Centrar el mapa en la media de las coordenadas
center = [unique_sites['Lat'].mean(), unique_sites['Long'].mean()]
site_map = folium.Map(location=center, zoom_start=4, tiles='CartoDB positron')

# Estilo visual
circle_radius = 1000  # en metros
text_style = "font-size: 12px; color:#d35400; font-weight:bold;"

# Agregar círculos y marcadores para cada sitio
for _, row in unique_sites.iterrows():
    lat = row['Lat']
    lon = row['Long']
    site_name = row['Launch Site']

    # Círculo que marca el área del sitio
    circle = folium.Circle(
        location=[lat, lon],
        radius=circle_radius,
        color='#d35400',
        fill=True,
        fill_opacity=0.3
    ).add_child(folium.Popup(site_name))

    # Marcador con texto personalizado
    marker = folium.map.Marker(
        [lat, lon],
        icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(0,0),
            html=f'<div style="{text_style}">{site_name}</div>'
        )
    )

    # Añadir ambos al mapa
    site_map.add_child(circle)
    site_map.add_child(marker)

# Mostrar el mapa interactivo
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# --- Librerías necesarias ---
import piplite
await piplite.install(['folium'])
await piplite.install(['pandas'])

import io
from js import fetch
import pandas as pd
import folium
from folium import Map, Marker, Popup
from folium.features import DivIcon
from folium.plugins import MousePosition
from math import sin, cos, sqrt, atan2, radians

# --- Función para calcular distancia entre coordenadas ---
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # radio aproximado de la Tierra en km
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# --- Cargar CSV remoto ---
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df = pd.read_csv(spacex_csv_file)

# Seleccionar columnas relevantes
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]

# Filtrar solo el sitio de lanzamiento VAFB SLC-4E
selected_site = 'VAFB SLC-4E'
launch_sites_df_filtered = spacex_df[spacex_df['Launch Site'] == selected_site].groupby('Launch Site', as_index=False).first()
spacex_df_filtered = spacex_df[spacex_df['Launch Site'] == selected_site]

# Coordenadas del punto de costa
coastline_point = [34.63319, -120.62628]

# Crear el mapa centrado en el sitio seleccionado con OpenStreetMap
site_center = [launch_sites_df_filtered['Lat'].mean(), launch_sites_df_filtered['Long'].mean()]
site_map = folium.Map(location=site_center, zoom_start=5, tiles='OpenStreetMap')

# --- Añadir MousePosition con formato ---
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)

# --- Añadir círculo y etiqueta del sitio ---
for _, row in launch_sites_df_filtered.iterrows():
    lat = row['Lat']
    lon = row['Long']
    site_name = row['Launch Site']
    
    circle = folium.Circle(
        location=[lat, lon],
        radius=1000,
        color='#d35400',
        fill=True,
        fill_opacity=0.3
    ).add_child(folium.Popup(site_name))
    
    label = folium.map.Marker(
        [lat, lon],
        icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(0,0),
            html=f'<div style="font-size:12px; color:#d35400; font-weight:bold;">{site_name}</div>'
        )
    )
    site_map.add_child(circle)
    site_map.add_child(label)

# --- Añadir PolyLine y marcador de distancia para cada lanzamiento ---
for _, row in spacex_df_filtered.iterrows():
    lat = row['Lat']
    lon = row['Long']
    
    # PolyLine desde lanzamiento hasta la costa
    line = folium.PolyLine(locations=[[lat, lon], coastline_point], color='blue', weight=1.5, opacity=0.7)
    site_map.add_child(line)
    
    # Calcular distancia
    distance_km = calculate_distance(lat, lon, coastline_point[0], coastline_point[1])
    
    # Marcador en la costa con distancia
    marker = folium.map.Marker(
        location=coastline_point,
        icon=DivIcon(
            icon_size=(250,36),
            icon_anchor=(0,0),
            html=f'<div style="font-size:12px; color:blue;"><b>{selected_site}: {distance_km:.2f} km</b></div>'
        )
    )
    site_map.add_child(marker)

# --- Mostrar mapa ---
site_map

```
